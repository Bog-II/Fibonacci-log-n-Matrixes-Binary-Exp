```python
import numpy as np
import time 

start = time.time()

def fib(n):
    
    if n == 0:
        return 0
    
    A = np.array([[0,1],[1,1]], dtype=np.object)
    res = np.array([[0,1]], dtype=np.object)

    n -= 1
    
    # binary exponention below
    while n > 0:
      if n % 2 == 1:
        res = res.dot(A)
      A = A.dot(A)
      n //= 2

    return res[0,1]

print(fib(10000))

end = time.time()
```

```python
print(end-start)
```

```python
array = [fib(n) for n in range(100)]
print(array)
```
